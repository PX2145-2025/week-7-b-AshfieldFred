```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import random, seed
```

```python
N=100
for i in range(0,N):
  steps=1000
  def walk(steps):
    x = np.random.choice([-1, 1], size=steps)
    x = np.cumsum(x)
    return x
  x=walk(steps)
  plt.plot(x)
plt.title('100 walks of 1000 steps')
plt.xlabel('Number of Steps')
plt.ylabel('Position')
plt.grid()
```

```python
n=10
particles = np.zeros((steps, n), int)
for i1 in range(n):
  x=walk(steps)
  particles[:,i1]=x

xmean = np.zeros(steps)
xmean_sq = np.zeros(steps)
for i2 in range (steps):
  xmean[i2] = np.mean(particles[i2,:])
  xmean_sq[i2] = np.mean(particles[i2,:]**2)

plt.plot(particles)
plt.title('10 random walks of 1000 steps')
plt.xlabel('Number of Steps')
plt.ylabel('Position')
plt.grid()
plt.show()

plt.plot(xmean)
plt.xlabel('Number of steps')
plt.ylabel('Mean Position')
plt.title('Mean position of N random walks at each step')
plt.grid()
plt.show() #The mean stays around zero for most of the time but as N gets large, it diverges away from 0

plt.plot(xmean_sq)
plt.xlabel('Number of steps')
plt.ylabel('Mean square position')
plt.title('Mean square position of N random walks at each step')
plt.grid()
plt.show() #The mean square position increases linearly
```
